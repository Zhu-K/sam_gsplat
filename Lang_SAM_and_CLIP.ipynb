```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
from lang_sam import LangSAM

image = "https://static01.nyt.com/images/2020/09/08/well/physed-cycle-walk/physed-cycle-walk-videoSixteenByNineJumbo1600-v2.jpg"
# image = "./image.jpg"
text_prompt = "person"

def download_image(url):
    response = requests.get(url)
    response.raise_for_status()
    return Image.open(BytesIO(response.content)).convert("RGB")

def save_mask(mask_np, filename):
    mask_image = Image.fromarray((mask_np * 255).astype(np.uint8))
    mask_image.save(filename)

def display_image_with_masks(image, masks):
    num_masks = len(masks)

    fig, axes = plt.subplots(1, num_masks + 1, figsize=(15, 5))
    axes[0].imshow(image)
    axes[0].set_title("Original Image")
    axes[0].axis('off')

    for i, mask_np in enumerate(masks):
        axes[i+1].imshow(mask_np, cmap='gray')
        axes[i+1].set_title(f"Mask {i+1}")
        axes[i+1].axis('off')

    plt.tight_layout()
    plt.show()

def display_image_with_boxes(image, boxes, logits):
    fig, ax = plt.subplots()
    ax.imshow(image)
    ax.set_title("Image with Bounding Boxes")
    ax.axis('off')

    for box, logit in zip(boxes, logits):
        x_min, y_min, x_max, y_max = box
        confidence_score = round(logit.item(), 2)  # Convert logit to a scalar before rounding
        box_width = x_max - x_min
        box_height = y_max - y_min

        # Draw bounding box
        rect = plt.Rectangle((x_min, y_min), box_width, box_height, fill=False, edgecolor='red', linewidth=2)
        ax.add_patch(rect)

        # Add confidence score as text
        ax.text(x_min, y_min, f"Confidence: {confidence_score}", fontsize=8, color='red', verticalalignment='top')

    plt.show()

def print_bounding_boxes(boxes):
    print("Bounding Boxes:")
    for i, box in enumerate(boxes):
        print(f"Box {i+1}: {box}")

def print_detected_phrases(phrases):
    print("\nDetected Phrases:")
    for i, phrase in enumerate(phrases):
        print(f"Phrase {i+1}: {phrase}")

def print_logits(logits):
    print("\nConfidence:")
    for i, logit in enumerate(logits):
        print(f"Logit {i+1}: {logit}")

def main():
    # Suppress warning messages
    warnings.filterwarnings("ignore")

    try:
        if image.startswith("http"):
            image_pil = download_image(image)
        else:
            image_pil = Image.open(image).convert("RGB")

        model = LangSAM()
        masks, boxes, phrases, logits = model.predict(image_pil, text_prompt)

        if len(masks) == 0:
            print(f"No objects of the '{text_prompt}' prompt detected in the image.")
        else:
            # Convert masks to numpy arrays
            masks_np = [mask.squeeze().cpu().numpy() for mask in masks]

            # Display the original image and masks side by side
            display_image_with_masks(image_pil, masks_np)

            # Display the image with bounding boxes and confidence scores
            display_image_with_boxes(image_pil, boxes, logits)

            # Save the masks
            for i, mask_np in enumerate(masks_np):
                mask_path = f"image_mask_{i+1}.png"
                save_mask(mask_np, mask_path)

            # Print the bounding boxes, phrases, and logits
            print_bounding_boxes(boxes)
            print_detected_phrases(phrases)
            print_logits(logits)

    except (requests.exceptions.RequestException, IOError) as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
# Load the CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = load("ViT-B/32", device=device)
```

```python
# Function to compute similarity scores
import clip
from torchvision.transforms.functional import to_pil_image
def compute_similarity(image, masks, query):
    image = preprocess(image).unsqueeze(0).to(device)  # Preprocess image
    text = clip.tokenize([query]).to(device)  # Tokenize text

    with torch.no_grad():
        image_features = model.encode_image(image)
        text_features = model.encode_text(text)

    # Normalize features
    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)

    # Compute similarity with the original image
    similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)
    print(f"Similarity with original image: {similarity.item():.2f}%")

    # Convert the preprocessed image back to PIL Image for masking
    image_for_masking = to_pil_image(image.squeeze(0).cpu())

    # Process each mask
    for i, mask in enumerate(masks):
        # Assuming mask is a binary mask, use it to create a masked version of the original image
        # masked_image = np.array(image) * np.expand_dims(mask, axis=-1)  # Apply mask
        # masked_image = Image.fromarray(masked_image.astype(np.uint8))
        # masked_image_preprocessed = preprocess(masked_image).unsqueeze(0).to(device)
        mask_pil = Image.fromarray(mask).resize(image_for_masking.size)
        mask_tensor = ToTensor()(mask_pil).to(device)
        masked_image_tensor = image * mask_tensor


        with torch.no_grad():
            masked_image_features = model.encode_image(masked_image_tensor)

        # Normalize features
        masked_image_features /= masked_image_features.norm(dim=-1, keepdim=True)

        # Compute similarity
        similarity = (100.0 * masked_image_features @ text_features.T).softmax(dim=-1)
        print(f"Similarity with mask {i}: {similarity.item():.2f}%")

# Example usage
url = "https://static01.nyt.com/images/2020/09/08/well/physed-cycle-walk/physed-cycle-walk-videoSixteenByNineJumbo1600-v2.jpg"  # Load your image
image = download_image(url)
masks = [np.array(Image.open("/content/drive/MyDrive/lang-segment-anything/image_mask_1.png")),  # Load your masks as numpy arrays
         np.array(Image.open("/content/drive/MyDrive/lang-segment-anything/image_mask_2.png")),
         np.array(Image.open("/content/drive/MyDrive/lang-segment-anything/image_mask_3.png")),
         np.array(Image.open("/content/drive/MyDrive/lang-segment-anything/image_mask_4.png")),
         np.array(Image.open("/content/drive/MyDrive/lang-segment-anything/image_mask_5.png")),
         np.array(Image.open("/content/drive/MyDrive/lang-segment-anything/image_mask_6.png")),
         np.array(Image.open("/content/drive/MyDrive/lang-segment-anything/image_mask_7.png"))]
query = "each person in the image"

compute_similarity(image, masks, query)
```
